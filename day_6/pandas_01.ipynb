## ***5.1 Dataset Parts***
### ***1. Head:***
>- ***head()*** command is used to show the first five rows of the dataset.
>- Write the dataset name "df" and type "."
>- After type ".", all commends of this library will be shown.
>- Select the "head()" word.
#### ***By default no. of rows:***
>- First five rows will be shown from index 0 to 4.
>- In python, index(serial no.) starts from 0.
#### ***Specific no. of rows:***
>- Type the number of rows you want to show in ().

```python
# head() command is used to show the first five rows of the dataset.
# by Default head() shows first five rows.
df.head()
```

```python
# For specific no. of rows, type the numbers of rows you want to show in ().

df.head(3)
```

### ***2. Tail:***

> - tail() command is used to show the last five rows of the dataset.

```python
df.tail()
```

### ***3. Shape:***

>- "shape" means look of the dataset.

>- Type "df.shape"

>- "shape" is used to show the total number of rows and columns in the dataset.

>- Always no. of rows is written first and second no. of columns.

>- output (87,3)

```python
df.shape
```

## ***Why we write some command with ( ) and some command without ( )?***
### Ans:

### ***Object:***
> - Commands Without ( ) is called ***Object***.
> - Commands without parenthesis () is an object of that function which can be passed as an argument and later be called by appending () to it.
#### ***For Example:***
> - hello refers to the function object

### ***Function:***
> - Commands With ( ) is called ***Funcion***.
> - function with parenthesis () is actually calling (invoke) that function
#### ***For Example:***
> - hello() calls the function itself.
