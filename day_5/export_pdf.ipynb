# ***How to convert Jupyter Notebook to PDF using Python***

## **Steps to convert Jupyter Notebook to PDF**

### **Step 1:**
#### Install the pdfkit library and jupyter notebook.
### **Step 2:**
#### click on three dots of your notebook and select **Export** option.

### **Step 3:**
#### Select **PDF** option.

### **Step 4:**
#### Save your notebook as PDF.
### **Step 5:**
#### Open your PDF in your browser.
